# Importing the required libraries and loading the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="white", color_codes=True)
```

```python
iris = pd.read_csv("IRIS.csv")
```

# Counting number of Species

```python
iris["species"].value_counts()
```

# visualizing the data

```python
sns.FacetGrid(iris, hue="species",height=8).map(plt.scatter,"sepal_length","petal_width").add_legend()
```

```python
sns.pairplot(iris, hue='species')
plt.title('Pair Plot - Iris Dataset')
plt.show()
```

```python
correlation_matrix = iris.drop('species', axis=1).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap - Iris Dataset')
plt.show()
```

# Converting the species to 0,1 and 2

```python
mapi ={'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}
iris["species"]= iris['species'].map(mapi)
```

```python
iris.head()
```

#  Giving values for x and y and Importing Logistic regression and training our model by fitting x and y thus calculating the accuracy score

```python
x=iris[['sepal_length','sepal_width','petal_length','petal_width']].values
```

```python
y=iris[['species']].values
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model =LogisticRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
expected = y
predicted = model.predict(x)
predicted
```

# Implementing Classinfication report and confusion matrix to understand our result and creating custom input to check our result

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(expected,predicted))
```

```python
print(metrics.confusion_matrix(expected,predicted))
```
