```python
reg=df['region'].value_counts().index[:5]
reg1=df['region_txt'].value_counts().index[:5]
dic={}
for each in range(0,5):
    dic[reg[each]]=reg1[each]
```

```python
def plot(area,name):
    met=df[df['region']==area]
    met.iyear.plot(kind='hist',bins=30,figsize=(10,5),color='red',fontsize='15')
    plt.xlabel('year',fontsize=15)
    plt.ylabel('Frequency',fontsize=15)
    plt.title(name)
    plt.show()
```

```python
for num in dic.keys():
    plot(num,dic[num]);
```
